```python
#1. ¿Cuál es el país más “feliz” del 2021 según la data?
pais2021 = happiness2021['Ladder score'].idxmax()
pais2021 = happiness2021.loc[pais2021, 'Country name']
print(pais2021)
```

```python
#2. ¿Cuál es el país más “feliz” del 2021 por continente según la data?
paises_por_region = happiness2021.groupby('Regional indicator')['Ladder score'].idxmax()
resultado = happiness2021.loc[paises_por_region, ['Regional indicator', 'Country name', 'Ladder score']].reset_index(drop=True)
print(resultado.to_string(index=False))
```

```python
#3. ¿Cuál es el país que más veces ocupó el primer lugar en todos los años?
paismasfelizporanio = happiness.groupby(['year', 'Country name'])['Life Ladder'].max()
paismasfeliz = paismasfelizporanio.groupby('year').idxmax()
print(paismasfelizporanio[paismasfeliz.values].to_string())
```

```python
#4. ¿Qué puesto de Felicidad tiene el país con mayor GDP del 2020?

#Creando csv con los registros del año 2020 y ordenado Life Ladder descendemente. Se le añadió la columna Happiness para tener los índices de felicidad.
happiness2020 = happiness[happiness["year"] == 2020]
happiness2020_sorted = happiness2020.sort_values(by=["Life Ladder"], ascending=False)[["Country name", "Log GDP per capita"]]
happiness2020_sorted = happiness2020_sorted.reset_index(drop=True)
happiness2020_sorted = happiness2020_sorted.assign(Happiness=range(1, len(happiness2020_sorted)+1))
happiness2020_sorted.to_csv("happiness2020_sorted.csv", index=False)

#Filtrando el archivo happiness2020_sorted.csv para saber el país y posición de índice de felicidad del país con mayor log GDP per capita.
happiness2020_GDP = pd.read_csv("happiness2020_sorted.csv")
highest_gdp = happiness2020_GDP[happiness2020_GDP['Log GDP per capita'] == happiness2020_GDP['Log GDP per capita'].max()]
resultado = highest_gdp[["Country name", "Happiness"]].reset_index(drop=True)
print(resultado.to_string(index=False))
```

```python
#5. ¿En que porcentaje a variado a nivel mundial el GDP promedio del 2020 respecto al 2021? ¿Aumentó o disminuyó?

#Calculando el promedio de GDP del año 2021.
average_log_gdp_2021 = happiness2021['Logged GDP per capita'].mean()

#Cambiando los valores null de la columna 'Log GDP per capita' y calculando el promedio de GDP del año 2020.
GDP2020 = pd.read_csv("happiness2020_sorted.csv")
mean_log_gdp_2020 = GDP2020['Log GDP per capita'].mean()
GDP2020['Log GDP per capita'].fillna(mean_log_gdp_2020, inplace=True)
average_log_gdp_2020 = GDP2020['Log GDP per capita'].mean()

#Comparando el promedio de GDP del año 2021 con el promedio de GDP del año 2020.
if average_log_gdp_2021 > average_log_gdp_2020:
    print("El promedio de GDP per capita en 2021 fue mayor que en 2020.")
elif average_log_gdp_2021 < average_log_gdp_2020:
    print("El promedio de GDP per capita en 2020 fue mayor que en 2021.")
else:
    print("El promedio de GDP per capita en 2020 y en 2021 fue igual.")

# Calculando el porcentaje de variación de GDP per capita entre el 2020 y el 2021
porcentaje = ((average_log_gdp_2021 - average_log_gdp_2020) / average_log_gdp_2020) * 100
print(f"{porcentaje:.2f}%")
```

```python
#6. ¿Cuál es el país con mayor expectativa de vide (“Healthy life expectancy at birth”)? Y ¿Cuánto tenia en ese indicador en el 2019?

#Buscando el país con mayor Healthy life expectancy at birth desde el 2005 al 2020.
life = happiness.groupby('year')['Healthy life expectancy at birth'].idxmax()
resultado = happiness.loc[life, ['Country name', 'Healthy life expectancy at birth', 'year']].reset_index(drop=True)
print(resultado)

print('')
print('País con mayor Healthy life expentancy del 2021:')
print('')

#Buscando el país con mayor Healthy life expectancy at birth en el 2021.
life2021 = happiness2021['Healthy life expectancy'].idxmax()
resultado = happiness2021.loc[life2021, ['Country name', 'Healthy life expectancy']]
print(resultado.to_string())
```
